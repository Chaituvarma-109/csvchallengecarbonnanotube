```python
ll = []
with open("carbon_nanotubes.csv", 'r') as cf:
    data_csv = csv.reader(cf, delimiter ='\n')
    for i in data_csv:
        ll.append(i[0].split(';'))
with open("new_carbon_nanotubes.csv", 'w', newline='') as ncf:
        wobj = csv.writer(ncf)
        wobj.writerows(ll)
```
